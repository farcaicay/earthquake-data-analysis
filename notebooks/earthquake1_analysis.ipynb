**4. Analisis Exploratorio del Dataset**

```python
df.info()
```

```python
df.describe()
```

```python
df["utc_date"] = pd.to_datetime(df["utc_date"])
```

```python
plt.hist(df["magnitude_m"], bins=20)
plt.title("Distribución de magnitudes")
plt.xlabel("Magnitud")
plt.ylabel("Cantidad de sismos")
plt.show()
```

```python
df["pais"].value_counts().plot(kind="bar")
plt.title("Sismos por Ubicación")
plt.xlabel("País")
plt.ylabel("Cantidad")
plt.show()
```

```python
df["year"] = df["utc_date"].dt.year

df["year"].value_counts().sort_index().plot()
plt.title("Sismos por año")
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.show()
```
